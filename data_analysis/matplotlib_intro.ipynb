```python
import scipy.stats as stats

# Generate sample data (1000 random values from normal distribution)
data = np.random.normal(50, 15, 1000)

# Calculate bins using different rules
sqrt_bins = int(np.sqrt(len(data)))
sturges_bins = int(1 + np.log2(len(data)))
rice_bins = int(2 * (len(data) ** (1/3)))

# Freedman-Diaconis Rule
q75, q25 = np.percentile(data, [75 ,25])
iqr = q75 - q25
fd_bin_width = 2 * iqr / (len(data) ** (1/3))
fd_bins = int((data.max() - data.min()) / fd_bin_width)

# Scotts Rule
scott_bin_width = 3.5 * np.std(data) / (len(data) ** (1/3))
scott_bins = int((data.max() - data.min()) / scott_bin_width)

# Plot histograms
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

bins_methods = [sqrt_bins, sturges_bins, rice_bins, fd_bins, scott_bins]
titles = ["Square Root Rule", "Sturges' Rule", "Rice Rule", "Freedman-Diaconis", "Scotts Rule"]

for ax, bins, title in zip(axes, bins_methods, titles):
    ax.hist(data, bins=bins, color='skyblue', edgecolor='black', alpha=0.7)
    ax.set_title(f"{title} (Bins = {bins})")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4.5,5,6]
y = [2,5,6,5,0.1,100]

plt.plot(x,y)

plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Simple Graph')
plt.legend('Legendary Line')

plt.show()
```

```python
# Creating a customized Line Text

x = [1,2,3,4.5,5,6]
y = [2,5,6,5,0.1,100]

plt.plot(x,y,color = 'red')
```
